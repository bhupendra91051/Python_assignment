Question 1: What is Simple Linear Regression?

Answer:  
Simple Linear Regression is a statistical method used to model the relationship between one independent variable and one dependent variable. It assumes that this relationship can be represented by a straight line. The model estimates how much the dependent variable changes when the independent variable changes by one unit.

The general equation is:
Y = β₀ + β₁X + ε,
where β₀ is the intercept, β₁ is the slope, and ε is the error term.

Question 2: What are the key assumptions of Simple Linear Regression?

Answer:  
The key assumptions of Simple Linear Regression are:

There is a linear relationship between the independent and dependent variables.

Observations are independent of each other.

The residuals are normally distributed.

Homoscedasticity exists, meaning constant variance of residuals.

There is no significant outlier that heavily influences the model.

Question 3: What is heteroscedasticity, and why is it important to address in regression models?

Answer:  
Heteroscedasticity occurs when the variance of the residuals is not constant across all levels of the independent variable. Instead, the spread of errors increases or decreases with the value of the predictor.

It is important to address heteroscedasticity because it can lead to inefficient estimates, biased standard errors, and unreliable hypothesis tests, which may result in incorrect conclusions.

Question 4: What is Multiple Linear Regression?

Answer:  
Multiple Linear Regression is an extension of Simple Linear Regression that models the relationship between one dependent variable and two or more independent variables. It helps understand how multiple factors simultaneously affect the outcome.

The general equation is:
Y = β₀ + β₁X₁ + β₂X₂ + ... + βₙXₙ + ε

Question 5: What is polynomial regression, and how does it differ from linear regression?

Answer:  
Polynomial Regression is a type of regression in which the relationship between the independent and dependent variables is modeled as an nth-degree polynomial. Unlike linear regression, it can capture non-linear relationships while still using linear model estimation techniques.

For example, a 2nd-degree polynomial regression includes squared terms of the predictor variable.

Question 6: Implement a Python program to fit a Simple Linear Regression model and plot the regression line.

Answer:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

model = LinearRegression()
model.fit(X, Y)

Y_pred = model.predict(X)

plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, Y_pred, color='red', label='Regression Line')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Simple Linear Regression")
plt.legend()
plt.show()

print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```

Question 7: Fit a Multiple Linear Regression model and check multicollinearity using VIF.

Answer:

Question 8: Implement polynomial regression (2nd degree) and plot the curve.

Answer:

```python
from sklearn.preprocessing import PolynomialFeatures

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

model = LinearRegression()
model.fit(X_poly, Y)

X_line = np.linspace(1, 5, 100).reshape(-1, 1)
X_line_poly = poly.transform(X_line)
Y_line = model.predict(X_line_poly)

plt.scatter(X, Y, color='blue')
plt.plot(X_line, Y_line, color='red')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("2nd Degree Polynomial Regression")
plt.show()
```

Question 9: Create a residuals plot and assess heteroscedasticity.

Answer:

```python
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

model = LinearRegression()
model.fit(X, Y)

Y_pred = model.predict(X)
residuals = Y - Y_pred

plt.scatter(X, residuals)
plt.axhline(0, color='red')
plt.xlabel("X")
plt.ylabel("Residuals")
plt.title("Residuals Plot")
plt.show()
```
