Question 1: What is hypothesis testing in statistics?

Answer:  
Hypothesis testing is a statistical method used to make decisions about a population parameter based on sample data. It involves making an assumption (called the null hypothesis) and then determining the likelihood of observing the sample data if that assumption were true. If the probability of the observed data under the null hypothesis is very low (below a predefined significance level), the null hypothesis is rejected in favor of an alternative hypothesis. Hypothesis testing is commonly used to validate claims or theories in scientific research.

Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

Answer:  
The null hypothesis (denoted as H₀) is a default assumption that there is no effect or no difference between groups or variables. It represents a position of skepticism or status quo.
The alternative hypothesis (denoted as H₁ or Ha) contradicts the null and proposes that there is an effect or a significant difference.

Difference:

H₀ : Assumes no change/effect (e.g., the mean is equal to a certain value).

H₁ : Assumes a change/effect exists (e.g., the mean is not equal to that value).
The testing process aims to gather evidence against H₀ to support H₁.

Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

Answer:  
The significance level (denoted as α) is the probability of rejecting the null hypothesis when it is actually true (Type I error). It defines the threshold for statistical significance. Common values for α are 0.05 (5%), 0.01 (1%), or 0.10 (10%).

Role in hypothesis testing:

- If the p-value is less than α, we reject the null hypothesis (evidence suggests a statistically significant effect).

- If the p-value is greater than α, we fail to reject the null hypothesis (insufficient evidence against H₀).
Thus, the significance level helps determine whether an observed result is due to chance or a real effect.

Question 4: What are Type I and Type II errors? Give examples of each.

Answer:

- Type I Error (False Positive) : Occurs when the null hypothesis is true, but we incorrectly reject it.
Example: A medical test wrongly indicating a disease when the person is healthy.

- Type II Error (False Negative) : Occurs when the null hypothesis is false, but we fail to reject it.
Example: A medical test failing to detect a disease when the person actually has it.

These errors represent the risks in decision-making based on sample data.  

Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.

Answer:

| **Test**     | **Z-Test**                                                   | **T-Test**                                                    |
|--------------|--------------------------------------------------------------|---------------------------------------------------------------|
| **Use**      | When population variance is known and/or sample size is large (n > 30) | When population variance is unknown and sample size is small (n ≤ 30) |
| **Distribution** | Standard normal distribution                                 | t-distribution                                                 |
| **Application**  | Comparing sample mean to population mean, or between two large samples | Small sample comparisons or when variance is unknown            |

**Summary:**  
Use a **Z-test** for large samples or when the population variance is known, and a **T-test** for small samples or when the population variance is unknown.



Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.  

(Include your Python code and output in the code box below.)
Hint: Generate random number using random function.

Answer:  


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10
p = 0.5

# Generate binomial data
data = np.random.binomial(n=n, p=p, size=1000)

# Plot histogram
plt.hist(data, bins=range(n+2), edgecolor='black', align='left')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.   

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
                        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
                        50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
                        50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
  
  (Include your Python code and output in the code box below.)    

 Answer:

```python
import numpy as np
from scipy import stats

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Convert to numpy array
data = np.array(sample_data)

# Hypothesis: H0: mu = 50
mu = 50
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)

# Z-statistic
z = (sample_mean - mu) / (sample_std / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

print(f"Z-statistic: {z:.4f}")
print(f"P-value: {p_value:.4f}")
```

Interpretation:

If p-value < 0.05 → reject H₀

If p-value > 0.05 → fail to reject H₀

Question 8: Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.   
(Include your Python code and output in the code box below.)   
Answer:

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Simulate normal data
np.random.seed(0)
data = np.random.normal(loc=100, scale=15, size=100)

# Calculate confidence interval
mean = np.mean(data)
std_err = stats.sem(data)
ci = stats.t.interval(0.95, len(data)-1, loc=mean, scale=std_err)

print(f"95% Confidence Interval: {ci}")

# Plot
plt.hist(data, bins=15, edgecolor='black', alpha=0.7)
plt.axvline(ci[0], color='red', linestyle='dashed', label='Lower CI')
plt.axvline(ci[1], color='green', linestyle='dashed', label='Upper CI')
plt.axvline(mean, color='blue', label='Mean')
plt.title("Normal Distribution with 95% Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.show()
```

Question 9: Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.   
(Include your Python code and output in the code box below.)   
Answer:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

# Generate sample data
data = np.random.normal(loc=70, scale=10, size=100)

# Calculate Z-scores
z_scores = zscore(data)

# Plot histogram
plt.hist(z_scores, bins=15, edgecolor='black', alpha=0.7)
plt.title('Histogram of Z-scores')
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

Explanation:  
Z-scores measure how many standard deviations a data point is from the mean:

Z = 0 → data point is at the mean

Z > 0 → data point is above the mean

Z < 0 → data point is below the mean
They are used in standardization and outlier detection.
