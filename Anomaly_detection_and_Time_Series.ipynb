Question 1: What is Dimensionality Reduction? Why is it important in Machine Learning?

Answer:  
Dimensionality reduction is a data preprocessing technique used to reduce the number of input variables or features in a dataset while retaining the most important information. It transforms high-dimensional data into a lower-dimensional representation.

Dimensionality reduction is important in machine learning because it:

Reduces computational complexity and training time

Helps prevent overfitting by removing redundant features

Improves model accuracy and performance

Enhances data visualization and interpretability

Mitigates the “curse of dimensionality”

Question 2: Name and briefly describe three common dimensionality reduction techniques.

Answer:  

Principal Component Analysis (PCA):
PCA is a statistical technique that converts correlated features into a smaller number of uncorrelated variables called principal components, capturing maximum variance.

Linear Discriminant Analysis (LDA):
LDA is a supervised dimensionality reduction method that maximizes class separability while reducing feature dimensions.

t-SNE (t-distributed Stochastic Neighbor Embedding):
t-SNE is a non-linear technique primarily used for data visualization in two or three dimensions.

Question 4: Explain the concept of anomaly detection and its significance.

Answer:  
Anomaly detection is the process of identifying unusual patterns or observations in data that deviate significantly from normal behavior.

Its significance lies in:

Fraud detection in banking and finance

Cybersecurity and intrusion detection

Fault detection in industrial systems

Medical diagnosis and monitoring

Question 5: List and briefly describe three types of anomaly detection techniques.

Answer:  

Statistical Methods:
These methods assume a statistical distribution and identify anomalies using measures like mean, variance, and z-score.

Distance-Based Methods:
Anomalies are detected based on their distance from neighboring data points.

Machine Learning-Based Methods:
Techniques such as Isolation Forest and One-Class SVM learn normal behavior and flag deviations as anomalies.

Question 6: What is time series analysis? Mention two key components of time series data.

Answer:  
Time series analysis involves analyzing data points collected at regular time intervals to identify patterns, trends, and future behavior.

Two key components of time series data are:

Trend: Long-term upward or downward movement

Seasonality: Regular repeating patterns over a fixed time period

Question 7: Describe the difference between seasonality and cyclic behavior in time series.

Answer:  
Seasonality refers to predictable and recurring patterns that occur at regular intervals, such as monthly or yearly sales increases during festivals.

Cyclic behavior refers to long-term fluctuations that do not have a fixed or predictable period, such as economic expansions and recessions.

Question 8: Write Python code to perform K-Means clustering on a sample dataset.

Answer:

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generate sample data
X, y = make_blobs(n_samples=200, centers=3, random_state=42)

# Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Plot the results
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis')
plt.title("K-Means Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

Question 9: What is inheritance in Object-Oriented Programming (OOP)? Provide a simple example in Python.

Answer:
Inheritance is an OOP concept in which a child class acquires the properties and behaviors of a parent class, promoting code reusability and hierarchical classification.

```python
class Vehicle:
    def start(self):
        print("Vehicle is starting")

class Car(Vehicle):
    def drive(self):
        print("Car is driving")

my_car = Car()
my_car.start()
my_car.drive()
```
