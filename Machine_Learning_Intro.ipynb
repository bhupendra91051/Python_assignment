Question 1: Explain the differences between AI, ML, Deep Learning (DL), and Data Science (DS).

Answer:  
Artificial Intelligence (AI) is the broad field that focuses on creating machines capable of performing tasks that normally require human intelligence, such as reasoning, decision-making, and problem-solving.

Machine Learning (ML) is a subset of AI that allows systems to learn patterns from data and improve performance without being explicitly programmed. It focuses on prediction and pattern recognition.

Deep Learning (DL) is a subset of machine learning that uses neural networks with multiple layers to model complex patterns. It is especially effective for tasks like image recognition, speech recognition, and natural language processing.

Data Science (DS) is an interdisciplinary field that uses statistics, programming, and domain knowledge to extract insights from data. It includes data collection, cleaning, analysis, visualization, and sometimes machine learning.

Question 2: What are the types of machine learning? Describe each with one real-world example.

Answer:  
Supervised learning uses labeled data to train models. The algorithm learns from input-output pairs.
Example: Email spam detection using labeled emails (spam or not spam).

Unsupervised learning works with unlabeled data to discover hidden patterns.
Example: Customer segmentation based on purchasing behavior.

Semi-supervised learning uses a small amount of labeled data with a large amount of unlabeled data.
Example: Image classification where only some images are labeled.

Reinforcement learning trains an agent to make decisions by rewarding correct actions.
Example: Game-playing AI like AlphaGo.

Question 3: Define overfitting, underfitting, and the bias-variance tradeoff in machine learning.

Answer:  
Overfitting occurs when a model learns noise and details from the training data too well, resulting in poor performance on new data.

Underfitting occurs when a model is too simple and fails to capture the underlying pattern of the data.

The bias-variance tradeoff represents the balance between a modelâ€™s simplicity and complexity. High bias leads to underfitting, while high variance leads to overfitting. A good model achieves a balance between the two.

Question 4: What are outliers in a dataset, and list three common techniques for handling them.

Answer:  
Outliers are data points that significantly differ from other observations in a dataset.

Three common techniques to handle outliers are:

Removing outliers using statistical thresholds such as Z-score or IQR.

Capping or flooring extreme values.

Transforming data using log or square root transformation.

Question 6: Write a Python program that creates a synthetic imbalanced dataset and prints the class distribution.

Answer:  

```python
from sklearn.datasets import make_classification
import numpy as np

X, y = make_classification(
    n_samples=1000,
    n_features=10,
    n_classes=2,
    weights=[0.9, 0.1],
    random_state=42
)

unique, counts = np.unique(y, return_counts=True)
print("Class distribution:")
for cls, count in zip(unique, counts):
    print(f"Class {cls}: {count}")
```

Question 8: Write a Python script to generate data, introduce missing values, impute them, and plot histograms.

Answer:

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(50, 10, 1000)

missing_indices = np.random.choice(1000, 50, replace=False)
data_with_missing = data.copy()
data_with_missing[missing_indices] = np.nan

plt.hist(data_with_missing[~np.isnan(data_with_missing)], bins=30)
plt.title("Before Imputation")
plt.show()

mean_value = np.nanmean(data_with_missing)
data_with_missing[np.isnan(data_with_missing)] = mean_value

plt.hist(data_with_missing, bins=30)
plt.title("After Imputation")
plt.show()
```

Question 9: Implement Min-Max scaling on [2, 5, 10, 15, 20] using sklearn and print the scaled array.

Answer:

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

data = np.array([2, 5, 10, 15, 20]).reshape(-1, 1)
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)

print(scaled_data)
```
