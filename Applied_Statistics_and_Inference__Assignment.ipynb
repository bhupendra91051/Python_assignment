Question 1: What are Type I and Type II errors in hypothesis testing, and how do they impact decision-making?

Answer:  
In hypothesis testing, errors can occur when making decisions based on sample data.

A Type I error occurs when the null hypothesis is true but is incorrectly rejected. This means we conclude that an effect or difference exists when it actually does not. The probability of making a Type I error is denoted by the significance level (α). Such errors may lead to false claims, unnecessary actions, or incorrect policy decisions.

A Type II error occurs when the null hypothesis is false but is not rejected. This means failing to detect a real effect. The probability of making a Type II error is denoted by β. This type of error can result in missed opportunities or failure to identify important effects.

In decision-making, reducing one type of error often increases the other, so a balance must be maintained depending on the situation.

Question 2: What is the P-value in hypothesis testing, and how should it be interpreted in the context of the null hypothesis?

Answer:  
The p-value is the probability of obtaining results as extreme as the observed data, assuming that the null hypothesis is true.

If the p-value is less than or equal to the chosen significance level (such as 0.05), the null hypothesis is rejected, indicating strong evidence against it. If the p-value is greater than the significance level, the null hypothesis is not rejected, indicating insufficient evidence to disprove it.

The p-value does not represent the probability that the null hypothesis is true; rather, it measures how compatible the data is with the null hypothesis.

Question 3: Explain the difference between a Z-test and a T-test, including when to use each.

Answer:  
A Z-test is used when the population standard deviation is known or when the sample size is large. It assumes that the sampling distribution of the mean follows a normal distribution.

A T-test is used when the population standard deviation is unknown and the sample size is small. It uses the t-distribution, which accounts for additional uncertainty caused by estimating the population standard deviation from the sample.

In practice, t-tests are more commonly used because population variance is rarely known.

Question 4: What is a confidence interval, and how does the margin of error influence its width and interpretation?

Answer:  
A confidence interval is a range of values used to estimate a population parameter, such as a mean. It provides an interval within which the true parameter is expected to lie with a certain level of confidence, such as 95%.

The margin of error determines the width of the confidence interval. A larger margin of error results in a wider interval, indicating less precision, while a smaller margin of error results in a narrower interval, indicating more precise estimates.

A higher confidence level or greater data variability increases the margin of error.

Question 5: Describe the purpose and assumptions of an ANOVA test. How does it extend hypothesis testing to more than two groups?

Answer:  
The purpose of an ANOVA (Analysis of Variance) test is to determine whether there are statistically significant differences among the means of three or more groups.

ANOVA assumes that observations are independent, data in each group is normally distributed, and variances across groups are equal.

ANOVA extends hypothesis testing beyond two groups by comparing the variation between group means to the variation within groups using a single test, thereby avoiding multiple pairwise comparisons that increase the risk of Type I errors.

Question 6: Write a Python program to perform a one-sample Z-test and interpret the result for a given dataset.

Answer:  

```python
import numpy as np
from scipy import stats

data = np.array([52, 55, 50, 53, 54])
population_mean = 50
population_std = 5

z = (np.mean(data) - population_mean) / (population_std / np.sqrt(len(data)))
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

print("Z-statistic:", z)
print("P-value:", p_value)
```

Question 8: Generate multiple samples from a non-normal distribution and implement the Central Limit Theorem using Python.

Answer:

```python
import numpy as np
import matplotlib.pyplot as plt

means = []

for _ in range(1000):
    sample = np.random.exponential(scale=2, size=50)
    means.append(np.mean(sample))

plt.hist(means, bins=30, edgecolor='black')
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.title("Central Limit Theorem Demonstration")
plt.show()
```

Question 9: Write a Python function to calculate and visualize the confidence interval for a sample mean.

Answer:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    n = len(data)
    z = norm.ppf(1 - (1 - confidence) / 2)
    margin = z * std / np.sqrt(n)
    return mean - margin, mean + margin

data = np.random.normal(50, 10, 100)
ci = confidence_interval(data)

plt.hist(data, bins=20)
plt.axvline(ci[0], color='red')
plt.axvline(ci[1], color='red')
plt.title("Confidence Interval for Sample Mean")
plt.show()
```

Question 10: Perform a Chi-square goodness-of-fit test using Python to compare observed and expected distributions, and explain the outcome.

Answer:

```python
from scipy.stats import chisquare

observed = [20, 30, 25, 25]
expected = [25, 25, 25, 25]

chi_stat, p_value = chisquare(observed, expected)

print("Chi-square statistic:", chi_stat)
print("P-value:", p_value)
```

Explanation:
If the p-value is less than 0.05, the null hypothesis is rejected, indicating a significant difference between observed and expected frequencies. If the p-value is greater than 0.05, there is no significant difference.
