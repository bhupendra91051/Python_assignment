Question 1: What is the difference between “Paste” and “Paste Special” in Excel? Briefly explain with examples.

Answer:  
In Microsoft Excel, the Paste command inserts all copied content such as values, formulas, formatting, and comments exactly as it appears in the source cell. It is used when an exact copy of the data is required.

Paste Special provides advanced options that allow users to paste specific elements like values only, formulas only, formats only, or transpose rows into columns.

Example:
If a cell contains the formula =A1+B1, using Paste will copy the formula, while Paste Special → Values will paste only the calculated result.

Question 2: Describe the functions and usefulness of “Freeze Panes” and “Split Panes” in Excel.

Answer:  
Freeze Panes is used to lock selected rows or columns so that they remain visible while scrolling through large datasets. This is helpful for keeping headings visible.

Split Panes divides a worksheet into multiple sections that can be scrolled independently. It is useful for comparing data in different parts of the same worksheet.

Question 3: Explain the difference between inserting a new row and inserting a new column in Excel. Can you insert multiple rows or columns at once?

Answer:  
Inserting a new row adds a horizontal line of cells and shifts existing rows downward. Inserting a new column adds a vertical line of cells and shifts existing columns to the right.

Yes, Excel allows inserting multiple rows or columns at the same time by selecting multiple rows or columns before choosing the insert option.

Question 5: Discuss the purpose of “XLOOKUP” and how it differs from the traditional “VLOOKUP” function.

Answer:  
XLOOKUP is used to search for values in a range and return corresponding results. It is more flexible and powerful than VLOOKUP.

Differences include:

XLOOKUP can search both left and right.

It does not require column index numbers.

It handles missing values better.

It replaces the limitations of VLOOKUP.
